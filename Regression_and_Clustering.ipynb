## Building a sentiment classifier

```python
products['rating'].show(view='Categorical')
```

```python
products = products[products['rating'] != 3]
```

```python
products.head()
```

```python
products['sentiments']=products['rating']>=4
```

```python
products.head()
```

```python
products['sentiments'].show(view='Categorical')
```

## Ready to Train Data

```python
train,test = products.random_split(.8,seed=0)
```

```python
model1 = graphlab.logistic_classifier.create(train, target='sentiments',features=['word_count'],validation_set=test)
```

## Evaluation of model

```python
model1.evaluate(test, metric='roc_curve')
```

```python
model1.show(view='Evaluation')
```
